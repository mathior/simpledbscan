# picapica - DBScan2D test set creation and validation

```python
import numpy as np
from sklearn.datasets.samples_generator import make_blobs, make_circles, make_moons
from matplotlib import pyplot as plt
```

```python
# Format (one point per line):
# x y label

def save(filename, xs, ys, labels):
    """Saves the lists xs, ys, and labels to a file.
    All lists must have equal lengths!"""
    assert(len(xs) == len(ys) and len(ys) == len(labels))
    with open(filename, 'w') as f:
        for i in range(len(xs)):
            f.write('%s %s %s\n' % (xs[i], ys[i], labels[i]))

def load(filename):
    """Loads a file where each line denotes a sample point in
    the format 'x y label' and returns a tuple (xs, ys, labels)."""
    with open(filename) as f:
        lines = [x.strip() for x in f.readlines() if x.strip()]
    values = [map(int, line.split()) for line in lines]
    return zip(*values)
```

```python
# In this case our dimension domain is integer and not scaled/normalized at the origin
# therefore some adjustments are necessary: convert sample point dimensions (independent)
# to int and move all into the first quadrant.
# Also increment label by one because in the clustering process a label 0 denotes an
# unassigned point (and therefore should not appear in the cluster labels when the
# process is finished)

def inc1(num_list):
    return [x + 1 for x in num_list]

def create_blobs_sample():
    centers = [[20, 20], [10, 20], [20, 10]]
    X, labels = make_blobs(n_samples=500, centers=centers, cluster_std=1.5, random_state=0)
    xs = [int(x[0] * 10) for x in X]
    ys = [int(x[1] * 10) for x in X]
    return xs, ys, inc1(labels)

def create_circle_sample():
    X, labels = make_circles(n_samples=200, shuffle=True, noise=0.03, random_state=None, factor=0.2)
    xs = [int(x[0] * 10 + 100) for x in X]
    ys = [int(x[1] * 10 + 100) for x in X]
    return xs, ys, inc1(labels)

def create_moon_sample():
    X, labels = make_moons(n_samples=200, shuffle=True, noise=0.05, random_state=None)
    xs = [int(x[0] * 15 + 100) for x in X]
    ys = [int(x[1] * 15 + 100) for x in X]
    return xs, ys, inc1(labels)


def plot_sample(xs, ys):
    plt.plot(xs, ys, 'o')
    plt.show()

def plot_labeled(xs, ys, labels):
    unique_labels = set(labels)
    colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))
    for x, y, label in zip(xs, ys, labels):
        if label == -1:  # Black used for noise.
            col = 'k'
        else:
            col = colors[label - 1]
        plt.plot(x, y, 'o', markerfacecolor=col, markeredgecolor='k', markersize=6)
    plt.show()
```

```python
save('blobs', *create_blobs_sample())
save('circles', *create_circle_sample())
save('moons', *create_moon_sample())
```

```python
blobs = load('blobs')
circles = load('circles')
moons = load('moons')

clustered_blobs = load('blobs_clustered')
clustered_circles = load('circles_clustered')
clustered_moons = load('moons_clustered')
```
